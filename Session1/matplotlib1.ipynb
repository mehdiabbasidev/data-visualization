```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 5, 7]
y2 = [1, 3, 5, 4, 6]
y3 = [0, 5 ,2 ,1, 8]

plt.plot(x, y1, label='Dataset 1', color='blue', ls='-', marker='o')
plt.plot(x, y2, label='Dataset 2', color='green', linestyle='-.', marker='s')
plt.plot(x, y3, label='Dataset 2', color='#a10000', linestyle='dotted', marker='>')

plt.title('Simple Line Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 5, 7]

font1 = {'family':'serif','color':'blue','size':20}
plt.plot(x, y, color='orange', linewidth=3, linestyle='--', marker='o', markersize=8)
plt.title('Line Style and Color Example',fontdict = font1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis',fontdict = font1)
plt.text(2,5,'Mehdi',color='yellow',bbox=dict(facecolor='black',edgecolor='blue',alpha=0.5)
# plt.grid(True)
plt.grid(axis='x',color="green")
plt.grid(axis='y',color="red")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.figure(figsize=(12,2))

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title('y = x²')

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='green')
plt.title('y = x³')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-2,2,100)
y=x**4
plt.plot(x,y,marker='>')
plt.show()
```

```python
import numpy as np
x=np.linspace(-2*np.pi,2*np.pi,100)

plt.figure(figsize=(12,3))

y=np.sin(x)
plt.subplot(1,2,1)
plt.plot(x,y)


y=np.cos(x)
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

# Pie

```python
labels=['ali','reza','mehdi','ahmad']
y=np.array([23,5,32,40])
colors=['green','yellow','black','pink']
plt.pie(y,colors=colors,labels=labels,startangle = 20)
plt.show()
```
