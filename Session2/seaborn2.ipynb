## Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
titanic.head()
# 7% random sample
sample_df = titanic.sample(frac=0.07, random_state=42)

plt.figure(figsize=(14,5))

sns.lineplot(
    data=sample_df,
    x="age",
    y="fare",
    hue="sex",
    style="class",
    markers=False,
    dashes=False
)
plt.title("Fare vs Age (Line Plot) - 5% Sample")
plt.show()

```

## Bar Plot

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=sample_df,
    x="class",
    y="fare",
    hue="sex",
    errorbar="sd",
    palette="Set2"
)

plt.title("Avg Fare by Class & Sex (Bar Plot)")
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=sample_df,
    x="age",
    y="fare",
    hue="sex",
    size="fare",
    sizes=(20, 200),
    alpha=0.9,
    palette="coolwarm"
)

plt.title("Age vs Fare (Scatter Plot)")
plt.show()
```

## Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=titanic,
    x="age",
    bins=30,
    kde=True,
    hue="sex",
    element="step",
    palette="husl"
)

plt.title("Age Distribution (Histogram)")
plt.show()
```

## Pie Chart

```python
import pandas as pd

survive_counts = titanic['sex'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    survive_counts,
    labels=survive_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("pastel")
)

plt.title("Gender Distribution (Pie Chart)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# ----- Load and Clean Dataset -----
sns.set_theme(style="whitegrid")  # Theme Ø­Ø±ÙÙ‡â€ŒØ§ÛŒ
titanic = sns.load_dataset("titanic")

df = titanic.dropna(subset=["age", "fare", "sex", "class"]).sample(120, random_state=0)

# ----- Create Figure -----
plt.figure(figsize=(20, 12))

# ===================== 1. Line Plot =====================
plt.subplot(2, 3, 1)
sns.lineplot(
    data=df,
    x="age",
    y="fare",
    hue="sex",
    style="class",
    markers=True,
    dashes=False,
    palette="Set2"
)
plt.title("Line Plot: Fare vs Age", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare ($)")


# ===================== 2. Bar Plot =====================
plt.subplot(2, 3, 2)
sns.barplot(
    data=df,
    x="class",
    y="fare",
    hue="sex",
    palette="Set3",
    errorbar=("sd")
)
plt.title("Bar Plot: Avg Fare by Class & Sex", fontsize=14)
plt.xlabel("Passenger Class")
plt.ylabel("Avg Fare ($)")


# ===================== 3. Scatter Plot =====================
plt.subplot(2, 3, 3)
sns.scatterplot(
    data=df,
    x="age",
    y="fare",
    hue="sex",
    size="fare",
    sizes=(20, 200),
    alpha=0.75,
    palette="coolwarm"
)
plt.title("Scatter Plot: Age vs Fare", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare ($)")


# ===================== 4. Histogram =====================
plt.subplot(2, 3, 4)
sns.histplot(
    data=df,
    x="age",
    hue="sex",
    kde=True,
    bins=25,
    element="step",
    palette="husl"
)
plt.title("Histogram: Age Distribution", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Count")


# ===================== 5. Pie Chart =====================
plt.subplot(2, 3, 5)

gender_counts = df["sex"].value_counts()
colors = sns.color_palette("pastel")

plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 12}
)
plt.title("Pie Chart: Gender Distribution", fontsize=14)


# ----- Layout -----
plt.tight_layout()
plt.show()
```
